**Adjusting the agegroups by calculation:**

Option: choose harmonised groups 18–34, 35–49, 50+ and approximate the splits of the 25–49 group using population proportions 
Population proportions are determined by population count on Eurostat:
https://ec.europa.eu/eurostat/databrowser/view/proj_19np__custom_17634975/default/table


Age Groups:
1. From 15 to 34 years
-	Include HO data 18 to 24 years and add %population from 15-17
-	Include HO data 25 to 49 years
    
o	split this value into age 25-34 = 12,0% of the population – 12,0% of commuters

2. From 35 to 49 years
-	Include HO data 25 to 49 years

o	split age 35-49 = 20,5% of the population – 20,5% of commuters

3. From 50 to 74 years
-	include HO data for 55 years and more and increase by %population between 50-54

Change different age groups:
1. From 15 to 34 years:

    Split group "25-49" into "25–34" and "35–49" based on population shares (row by row)

2. From 35 to 49 years:

    Increase percentage of group 18-34 by population percentage of 15-17 to change to "15-34"

3. From 50 to 74 years:

    Determine percentage 50-54 in population and adjust HO data "55+ years" so we have 50-55+ years (add all 55+ data) (assuming that 55+ will not be older than 74 years)

```python
# Define your population shares
pop_25_34 = 12.0
pop_35_49 = 20.5
# Calculate the total population share for the age groups "25-34" and "35-49"
#Add their population shares together to get the total:
#total_share = 12.0 + 20.5 = 32.5
total_share = pop_25_34 + pop_35_49

#Divide each group’s population share by the total to get their portion:
#share_25_34 = 12.0 / 32.5 ≈ 0.369 → 36.9%
#share_35_49 = 20.5 / 32.5 ≈ 0.631 → 63.1%
share_25_34 = pop_25_34 / total_share
share_35_49 = pop_35_49 / total_share

# Step 1: Select all rows with age_group "25-49"
boolean = df_home_office["age_group"] == "25-49"
rows_to_split = df_home_office[boolean].copy()

# Step 2: Create the two split rows
rows_25_34 = rows_to_split.copy()
rows_25_34["age_group"] = "25–34"
rows_25_34["2019"] = pd.to_numeric(rows_25_34["2019"], errors="coerce") * share_25_34

rows_35_49 = rows_to_split.copy()
rows_35_49["age_group"] = "35–49"
rows_35_49["2019"] = pd.to_numeric(rows_35_49["2019"], errors="coerce") * share_35_49

# Step 3: Drop original "25-49" rows and append new ones
df_home_office = df_home_office[~boolean] # Remove the original "25-49" rows ~means "not"
df_home_office = pd.concat([df_home_office, rows_25_34, rows_35_49], ignore_index=True)
```

```python
#Adjust group "18-34" based on population shares - 3.1% of the total population is 15-17 years old -- change to "15-24" for consistency with commuting data
#change object to numeric type to perform calculations
df_home_office["2019"] = pd.to_numeric(df_home_office["2019"], errors="coerce")

df_home_office.loc[df_home_office["age_group"] == "18-24", "2019"] *= (1 + 0.031)
df_home_office.loc[df_home_office["age_group"] == "18-24", "age_group"] = "15-24"
```

```python
df_home_office
```

```python
#Adjust group "55+" based on population shares - 7.6% of the total population is 50-54 years old -- change to "50-74" for consistency with commuting data

df_home_office["2019"] = pd.to_numeric(df_home_office["2019"], errors="coerce")

df_home_office.loc[df_home_office["age_group"] == "55+", "2019"] *= (1 + 0.076)
df_home_office.loc[df_home_office["age_group"] == "55+", "age_group"] = "50-74"
```

```python
df_home_office.age_group.unique()
```

```python
#Combine "15-24" and "25–34" into "15–34"

# First, make sure your '2019' column is numeric
df_home_office["2019"] = pd.to_numeric(df_home_office["2019"], errors="coerce")

# Filter the rows you want to combine
filtered_df = df_home_office["age_group"].isin(["15-24", "25–34"])
combined_rows = df_home_office[filtered_df].copy()

# Group by other relevant columns (like region, gender, etc.) and sum the 2019 values
combined = combined_rows.groupby(["region", "gender", "children"], as_index=False)["2019"].sum()

# Add new age_group label
combined["age_group"] = "15–34"

# Remove the old rows from the main DataFrame
df_home_office = df_home_office[~filtered_df]

# Append the new combined row
df_home_office = pd.concat([df_home_office, combined], ignore_index=True)
```
